## Zusammenführen aller Kandidierenden

```python
main_folder_path = '/content/drive/MyDrive/Projekt_Mobilisierung/Data/LINKE/Kandidierende'

dfs = []

for folder in os.listdir(main_folder_path):
    folder_path = os.path.join(main_folder_path, folder)

    if os.path.isdir(folder_path):
        csv_file_path = os.path.join(folder_path, f'{folder}_posts_01_04-08_10_ocr.csv')
        if os.path.exists(csv_file_path):
            df = pd.read_csv(csv_file_path)
            df['folder_id'] = folder
            dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)

final_df.to_csv('/content/drive/MyDrive/Projekt_Mobilisierung/Data/LINKE/Kandidierende/all_candidates_combined.csv', index=False)

final_df
```

```python
additional_csv_paths = [
    "/content/drive/MyDrive/Projekt_Mobilisierung/Data/LINKE/LINKE-Landtagsfraktion/linke_lt_posts_01_04-08_10_ocr.csv",
    "/content/drive/MyDrive/Projekt_Mobilisierung/Data/LINKE/LINKE-Parteiacc/linke_posts_01_04-08_10_ocr.csv"
]

for csv_path in additional_csv_paths:
    if os.path.exists(csv_path):
        additional_df = pd.read_csv(csv_path)
        folder_name = os.path.basename(os.path.dirname(csv_path))
        additional_df['folder_id'] = folder_name
        final_df = pd.concat([final_df, additional_df], ignore_index=True)

final_df.to_csv('/content/drive/MyDrive/Projekt_Mobilisierung/Data/LINKE/all_linke_accs.csv', index=False)

final_df
```

## Zusammenführen aller Accounts

```python
dfs = []

main_folder_path = '/content/drive/MyDrive/Projekt_Mobilisierung/Data'

for folder in ['SPD', 'CSU', 'GRUENE', 'AFD', 'FW', 'LINKE', 'FDP']:
    folder_path = os.path.join(main_folder_path, folder)
    if os.path.isdir(folder_path):
        csv_file_path = os.path.join(folder_path, f'all_{folder.lower()}_accs.csv')
        if os.path.exists(csv_file_path):
            df = pd.read_csv(csv_file_path)
            df['party'] = folder
            dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)

final_df.to_csv('/content/drive/MyDrive/Projekt_Mobilisierung/all_accs_combined.csv', index=False)
```

## Erstellung Annotationssample

```python
final_df = pd.read_csv('/content/drive/MyDrive/Projekt_Mobilisierung/Data/all_accs_combined.csv')

parteiacc_values = ['FDP-Parteiacc', 'SPD-Parteiacc', 'CSU-Parteiacc', 'AFD-Parteiacc', 'FW-Parteiacc', 'LINKE-Parteiacc', 'GRUNE-Parteiacc']
landtagsfraktion_values = ['CSU-Landtagsfraktion', 'AFD-Landtagsfraktion', 'SPD-Landtagsfraktion', 'GRUNE-Landtagsfraktion', 'FDP-Landtagsfraktion', 'FW-Landtagsfraktion']

def assign_value(folder_id):
    if folder_id in parteiacc_values:
        return "Parteiaccount"
    elif folder_id in landtagsfraktion_values:
        return "Fraktionsaccount"
    else:
        return "Kandi-Account"

final_df['acc_type'] = final_df['folder_id'].apply(assign_value)

final_df.to_csv('/content/drive/MyDrive/Projekt_Mobilisierung/Data/all_accs_combined.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

final_df = pd.read_csv('/content/drive/MyDrive/Projekt_Mobilisierung/Data/all_accs_combined.csv')

final_df['party_type_folder'] = (
    final_df['party'].astype(str) + '_' +
    final_df['type'].astype(str) + '_' +
    final_df['acc_type'].astype(str) + '_'
)

counts = final_df['party_type_folder'].value_counts()
valid_combinations = counts[counts > 1].index
final_df = final_df[final_df['party_type_folder'].isin(valid_combinations)]

sample_size = 1000

stratified_sample, _ = train_test_split(final_df, train_size=sample_size, stratify=final_df['party_type_folder'], random_state=42)

stratified_sample = stratified_sample.drop(columns=['party_type_folder'])

print(stratified_sample['party'].value_counts(normalize=True))
print(stratified_sample['type'].value_counts(normalize=True))
print(stratified_sample['acc_type'].value_counts(normalize=True))
print(final_df['party'].value_counts(normalize=True))
print(final_df['type'].value_counts(normalize=True))
print(final_df['acc_type'].value_counts(normalize=True))
```

## Hinzufügen Interaktionszahlen


```python
df1_path = '/content/drive/MyDrive/Projekt_Mobilisierung/Data/class_final_df_pm.csv'

df1 = pd.read_csv(df1_path)

df1.rename(columns={'Follower': 'follower'}, inplace=True)

df1['follower'] = pd.to_numeric(df1['follower'], errors='coerce')
df1['likes'] = pd.to_numeric(df1['likes'], errors='coerce')
df1['comments'] = pd.to_numeric(df1['comments'], errors='coerce')

df1['interaction_rate'] = (df1['likes'] + df1['comments']) / df1['follower']

df1.to_csv(df1_path, index=False)

print("Berechnung der Interaktionsrate abgeschlossen und Datei gespeichert.")
```
