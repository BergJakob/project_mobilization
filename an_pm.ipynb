# Analyse

```python
from google.colab import drive
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/Projekt_Mobilisierung/Data/all_posts.csv'
final_df = pd.read_csv(file_path)
```

```python
dfs = []

main_folder_path = '/content/drive/MyDrive/Projekt_Mobilisierung'

for folder in ['SPD', 'CDU', 'GRUENE', 'AFD', 'FDP', 'LINKE']:
    folder_path = os.path.join(main_folder_path, folder)
    if os.path.isdir(folder_path):
        csv_file_path = os.path.join(folder_path, f'all_{folder.lower()}_accs.csv')
        if os.path.exists(csv_file_path):
            df = pd.read_csv(csv_file_path)
            df['party'] = folder
            dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)

final_df.to_csv('/content/drive/MyDrive/Projekt_Mobilisierung/all_accs_combined.csv', index=False)
```

## Deskriptive Statistik

```python
final_df
```

### Posts

```python
party_post_counts = final_df.groupby('party').size().reset_index(name='post_count')
print(party_post_counts)

media_counts = final_df.groupby('party')['num_media'].sum().reset_index(name='total_media')

print(media_counts)
```

### Interaktionen

Likes

```python
likes_df = final_df.groupby('party')['num_likes'].sum().reset_index()

total_likes = final_df['num_likes'].sum()

likes_df = likes_df.append({'party': 'Gesamt', 'num_likes': total_likes}, ignore_index=True)

print(likes_df)

average_likes_per_post = final_df.groupby('party')['num_likes'].mean().reset_index(name='average_likes_per_post')

print(average_likes_per_post)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
bar_colors = ['#009EE0', '#005538', '#E3000F', 'grey'] # CSU, FW, AFD, GRÜNE, SPD, FDP, Linke  ['#black', '#f49404', '#009EE0', '#005538', '#E3000F', '#ffed00', '#6f003c','grey']
plt.bar(likes_df['party'], likes_df['num_likes'], color=bar_colors)
plt.xlabel('Partei')
plt.ylabel('Anzahl der Likes')
plt.title('Likes pro Partei')

plt.yticks(range(0, 4000001, 1000000))

plt.xticks(likes_df['party'], ['AFD', 'B90/Grüne', 'SPD', 'Gesamt'])

formatter = FuncFormatter(lambda x, _: f'{x / 1e6:.0f}M' if x >= 1e6 else f'{x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

for i, v in enumerate(likes_df['num_likes']):
    formatted_number = '{:,.0f}'.format(v).replace(',', '.')
    plt.text(i, v + 100000, formatted_number, color='black', ha='center', va='bottom')

plt.show()
```

Kommentare

```python
comments_df = final_df.groupby('party')['num_comments'].sum().reset_index()

total_comments = final_df['num_comments'].sum()

comments_df = comments_df.append({'party': 'Gesamt', 'num_comments': total_comments}, ignore_index=True)

print(comments_df)

average_comments_per_post = final_df.groupby('party')['num_comments'].mean().reset_index(name='average_comments_per_post')

print(average_comments_per_post)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
bar_colors = ['#009EE0', '#005538', '#E3000F', 'grey'] # CSU, FW, AFD, GRÜNE, SPD, FDP, Linke  ['#black', '#f49404', '#009EE0', '#005538', '#E3000F', '#ffed00', '#6f003c','grey']
bars = plt.bar(comments_df['party'], comments_df['num_comments'], color=bar_colors)
plt.xlabel('Partei')
plt.ylabel('Anzahl der Kommentare')
plt.title('Kommentare pro Partei')

plt.yticks(range(0, 150000, 10000))

plt.xticks(comments_df['party'], ['AFD', 'B90/Grüne', 'SPD', 'Gesamt'])

formatter = FuncFormatter(lambda x, _: f'{x / 1e6:.0f}M' if x >= 1e6 else f'{x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

for bar in bars:
    yval = bar.get_height()
    formatted_value = f'{int(yval):,}'.replace(',', '.')
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5000, formatted_value, ha='center', va='bottom', color='black')

plt.show()
```

## Hashtag-Analyse

```python
import pandas as pd
from collections import Counter

# Alle Hashtags aus der Spalte 'hashtags' extrahieren und in Kleinbuchstaben umwandeln
all_hashtags = [tag.lower() for tag in ','.join(final_df['hashtags'].dropna()).split(',')]

# Verwende ein Set, um doppelte Einträge zu entfernen, und zähle die Häufigkeit der Hashtags
hashtags_counter = Counter(all_hashtags)

# Die 20 häufigsten Hashtags erhalten
top_n_hashtags = hashtags_counter.most_common(20)
print(top_n_hashtags)
```

```python
import pandas as pd
from collections import Counter

all_hashtags = ','.join(final_df['hashtags'].dropna()).split(',')

hashtags_counter = Counter(all_hashtags)

hashtags_df = pd.DataFrame({'Hashtag': list(hashtags_counter.keys()),
                            'Häufigkeit': list(hashtags_counter.values())})

hashtags_df.sort_values(by='Häufigkeit', ascending=False, inplace=True)
hashtags_df.reset_index(drop=True, inplace=True)
hashtags_df.index += 1  # Starten Sie den Rang bei 1
hashtags_df.index.name = 'Rang'

df = hashtags_df.head(75)
df
```

```python
top_75_hashtags = df['Hashtag'][:75]
top_75_häufigkeiten = df['Häufigkeit'][:75]

plt.figure(figsize=(8, 12))
bars = plt.barh(top_75_hashtags, top_75_häufigkeiten, color='blue')

plt.xlabel('Häufigkeit', fontsize=12)
plt.ylabel('Hashtags', fontsize=12)
plt.title('Hashtags TOP 75 nach Häufigkeit', fontsize=14)

plt.yticks(range(len(top_75_hashtags)), ['#' + tag for tag in top_75_hashtags], fontsize=10)

plt.xticks(fontsize=10)

plt.tight_layout()
plt.show()
```
